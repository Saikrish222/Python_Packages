### Matplotlib Line plot

```python
# importing matplotlib module 
from matplotlib import pyplot as plt
x = [2,3,7,2,13,7]
y = [4,6,8,3,6,7]

plt.plot(x, y)


plt.show()
```

```python
from matplotlib import pyplot as plt
x = [2,3,7,-2,13,7]
y = [4,-6,8,-3,6,7]

plt.plot(x, y)


plt.show()
```

### Matplotlib bar plot

```python
from matplotlib import pyplot as plt
x = [1,3,5,7,6,4,2]
y = [2,4,6,7,5,3,1]

plt.bar(x,y)

plt.show()
```

```python
from matplotlib import pyplot as plt
x = [1,3,5,-7,6,4,2]
y = [-2,4,6,7,-5,3,1]

plt.bar(x,y)

plt.show()
```

### Matplotlib Histogram

```python
from matplotlib import pyplot as plt
x = [1,3,5,7,6,4,2]


plt.hist(x)

plt.show()
```

### Matplotlib Scatter Plot

```python
# importing matplotlib module 
from matplotlib import pyplot as plt


x = [4,6,3,9,4,6,2,5]

y = [4,4,6,7,4,9,2,5]

plt.scatter(x, y)

plt.show()
```

### Matplotlib PieChart

```python
from matplotlib import pyplot as plt

labels =['Batting','Bolwing','fielding']
sizes =[80,60,70]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Player Stats')
plt.show()
```

### Matplotlib Area Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1, y2 = [10, 20, 15, 25, 30], [5, 15, 10, 20, 25]

plt.fill_between(x, y1, y2, color='red', alpha=0.4, label='Area 1-2')
plt.plot(x, y1, label='Line 1', marker='o')
plt.plot(x, y2, label='Line 2', marker='o')


plt.xlabel('X-axis'), plt.ylabel('Y-axis'), plt.title('Area Chart')

plt.legend(), plt.show()
```
